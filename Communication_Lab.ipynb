```python
n = 10  # no:of symbols
tm = 100  # length of symbol

m = np.random.randint(0, 2, n)
# print(m)
even = m[0:n:2]
# print(even)
even_bits = 2*np.repeat(even,tm*2)-1
# print(even_bits)

odd = m[1:n:2]
# print(odd)
odd_bits=2*np.repeat(odd,tm*2)-1
# print(odd_bits)

A = 1
f = 1
# W = 100
t = np.linspace(0, 6*np.pi, (tm*n))
carrier=[A*np.cos(2*np.pi*f*t), A*np.sin(2*np.pi*f*t)]

product_even=[]
product_odd=[]
for t in range((tm*n)):
    product_even.append(even_bits[t]*carrier[0][t])
    product_odd.append(odd_bits[t]*carrier[1][t])

qpsk_even = product_even
qpsk_odd = product_odd
qpsk=[qpsk_odd[i] + qpsk_even[i] for i in range(tm*n)]

# plt.plot(qpsk_even)
# plt.show()
# plt.plot(qpsk_odd)
# plt.show()

plt.subplot(311)
plt.plot(np.repeat(m,tm))
plt.title("Message Signal")
plt.subplot(312)
plt.plot(carrier[0],label="Carrier 1 (cosine)")
plt.plot(carrier[1], label="Carrier 2 (sine)")
plt.title("Carrier Signals")
plt.legend()
plt.subplot(313)
plt.plot(qpsk, "r")
plt.title("QPSK Waveform")
plt.tight_layout()
plt.show()
```

```python
#Monte Carlo Simulation
N_bits = 100000
bits = np.random.randint(0, 2, N_bits) #QPSK signal

# QPSK mapping
x_re = 2*bits[::2] - 1  # even bits mapped to real part
x_im = 2*bits[1::2] - 1  # odd bits mapped to imaginary part
mod = x_re + 1j*x_im
#print(mod)

ber = []
snr = np.arange(start=-4,stop=10, step=2)
No = 1/(10**(snr/10))
sigma = np.sqrt(No/2) # Finding Noise
#print(sigma)

for i in range(len(sigma)):
    n = sigma[i] * (np.random.randn(N_bits//2) + 1j*np.random.randn(N_bits//2))
    y = mod + n  # received symbols

    # Demodulation
    y_re = np.real(y)
    y_im = np.imag(y)

    bits_hat = np.zeros(N_bits)
    bits_hat[::2] = 0.5*(1 + np.sign(y_re))
    bits_hat[1::2] = 0.5*(1 + np.sign(y_im))

    # Compute bit error rate
    ber.append(sum(bits != bits_hat) / N_bits)
    #print(berr)

ber_theory = 0.5 * erfc(np.sqrt(10**(snr/10)))

plt.semilogy(snr,ber_theory,"ro-",label="Theoretical BER")
plt.semilogy(snr,ber,"bo-", label="Practical BER")
plt.title("BER vs SNR for QPSK over AWGN Channel")
plt.xlabel("SNR")
plt.ylabel("BER")
plt.grid()
plt.legend()
plt.show()
```

To generate a string of message bits for all odd numbers up to 7 for at least 5 cycles and simulate the transmission through an AWGN (Additive White Gaussian Noise) channel.

```python
import numpy as np
import matplotlib.pyplot as plt

def generate_message():
  message = []
  for i in range(8):
    if i%2 != 0:
      # print(i,bin(i)[2:])
      message += bin(i)[2:]
  for i in range(len(message)):
    message[i] = int(message[i])
  return message

def AWGN_channel(message, snr_db, no_of_cycles):
  no_of_bits = len(message)
  # print("no:of bits:", no_of_bits)
  snr = 10**(snr_db/10)
  # print("SNR: ", snr)
  noise_power = 1.0 / snr
  # print("noise_power: ", noise_power)
  root_noise_power = np.sqrt(noise_power)
  # print("root_noise_power: ", root_noise_power)

  message = np.array(message)
  no_of_errors = 0
  Noise = []
  Rx = []

  for i in range(no_of_cycles):
    print("\nCycle",i+1)
    noise = np.random.normal(0,root_noise_power,no_of_bits)
    Noise.extend(noise)

    print("message: ", message)
    print("noise: ", noise)

    Tx_signal = message + noise
    print("Transmitted signal:", Tx_signal)

    Rx_signal = [1 if bit > 0.5 else 0 for bit in Tx_signal]
    Rx += Rx_signal
    print("Received signal:", Rx_signal)

    no_of_errors += sum(message[i] != Rx_signal[i] for i in range(no_of_bits))

  return no_of_errors, Noise, Rx

message_bits = generate_message()
snr_dB = 10
no_of_cycles = 5

no_of_errors, Noise, Rx = AWGN_channel(message_bits, snr_dB, no_of_cycles)

fig, a = plt.subplots(3,1)
t = np.linspace(0,50,45)

a[0].stem(t, message_bits*5)
a[0].set_title("Message")
a[1].plot(t, Noise)
a[1].set_title("Noise signal (AWGN Channel)")
a[2].stem(t, Rx)
a[2].set_title("Received signal")
plt.tight_layout()

print("\n")
print("Number of bit errors:", no_of_errors)
```
